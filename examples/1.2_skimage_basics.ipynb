### Imports

````python
from skimage import io
from skimage import transform
from skimage import img_as_ubyte

from skimage.color import rgb2gray
from skimage.color import rgba2rgb

import matplotlib.pyplot as plt
````

### Lendo e inspecionando a imagem

<img src="img_rep.png"/>

````python
img_path = 'sample_images/machine-learning-meetup.png'
````

````python
img = io.imread(img_path)
````

### Verificando shape (width, height, channels)
Note que agora nos temos 4 canais (RGBA)

````python
print(img.shape)
````

### Verificando tipo de dados

````python
print(img.dtype)
````

### Mostrando a imagem
Note que as cores estão corretas. Isso é porque a imagem é carregada como RGBA.

````python
plt.imshow(img)
````

### Convertendo para RGB

````python
img_rgb = rgba2rgb(img)
````

````python
plt.imshow(img_rgb)
````

````python
print(img_rgb.shape)
````

Note que o tipo de dados mudou para float64. Isso faz diferença?

````python
print(img_rgb.dtype)
````

Se nós olharmos para a média dos pixels, podemos ver uma diferença significativa. Lembram dos valores com o OpenCV?

````python
print('RGB - float64')
print('R:', round(img_rgb[:,:,0].mean(), 2))
print('G:', round(img_rgb[:,:,1].mean(), 2))
print('B:', round(img_rgb[:,:,2].mean(), 2))
````

### Convertendo para uint8
Após convertermos para uint8, podemos ver os mesmos resultados que o OpenCV.

````python
img_rgb_uint8 = img_as_ubyte(img)
````

````python
print('RGB - uint8')
print('R:', round(img_rgb_uint8[:,:,0].mean(), 2))
print('G:', round(img_rgb_uint8[:,:,1].mean(), 2))
print('B:', round(img_rgb_uint8[:,:,2].mean(), 2))
````

````python
plt.imshow(img_rgb_uint8)
````

### Redimensionando a imagem

````python
img_rgb_resized = transform.resize(img_rgb, (300, 300))
````

````python
plt.imshow(img_rgb_resized)
````

````python
print(img_rgb_resized.shape)
````

### Convertendo para tons de cinza

````python
img_gray = rgb2gray(img_rgb)
````

````python
print(img_gray.shape)
````

````python
plt.imshow(img_gray, cmap='gray')
````

### Lendo em tons de cinza diretamente

````python
img_gray2 = io.imread(img_path, as_gray=True)
````

````python
print(img_gray2.shape)
````

````python
plt.imshow(img_gray2, cmap='gray')
````
