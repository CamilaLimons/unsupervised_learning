### Imports

````python
import cv2
import matplotlib.pyplot as plt
````

### Lendo e inspecionando a imagem

<img src="img_rep.png"/>

````python
img_path = 'sample_images/machine-learning-meetup.png'
````

````python
img = cv2.imread(img_path)
````

### Verificando shape (width, height, channels)

````python
print(img.shape)
````

### Verificando tipo de dados

````python
print(img.dtype)
````

### Mostrando a imagem
Note que as cores não estão corretas. Isso é porque o OpenCV carrega as imagens em BGR por padrão.

````python
plt.imshow(img)
````

### Trocando a ordem dos canais

````python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
````

````python
plt.imshow(img_rgb)
````

### Verificando diferenças entre BGR e RGB

````python
print('BGR')
print('B:', round(img[:,:,0].mean(), 2))
print('G:', round(img[:,:,1].mean(), 2))
print('R:', round(img[:,:,2].mean(), 2))
````

````python
print('RGB')
print('R:', round(img_rgb[:,:,0].mean(), 2))
print('G:', round(img_rgb[:,:,1].mean(), 2))
print('B:', round(img_rgb[:,:,2].mean(), 2))
````

### Redimensionando a imagem

````python
img_rgb_resized = cv2.resize(img_rgb, (300, 300))
````

````python
plt.imshow(img_rgb_resized)
````

````python
print(img_rgb_resized.shape)
````

### Convertendo para tons de cinza

````python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
````

````python
print(img_gray.shape)
````

````python
plt.imshow(img_gray, cmap='gray')
````
