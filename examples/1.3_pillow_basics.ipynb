### Imports

````python
from PIL import Image

import numpy as np
import matplotlib.pyplot as plt
````

### Lendo e inspecionando a imagem

<img src="img_rep.png"/>

````python
img_path = 'sample_images/machine-learning-meetup.png'
````

````python
img = Image.open(img_path)
````

### Verificando shape (width, height, channels)

Note que quando utilizamos o método open do Pillow, nós não obtemos um array Numpy.

````python
print(img.shape)
````

Nós temos um objeto Pillow

````python
print(img)
````

### Convertendo para array Numpy

````python
img_array = np.asarray(img)
````

### Verificando shape (width, height, channels)

Nós podemos agora verificar o shape, mas note que novamente estamos obtendo um quarto canal (RGBA).

````python
print(img_array.shape)
````

Mas o tipo é uint8

````python
print(img_array.dtype)
````

### Mostrando a image
Funciona com o objeto original `img`

````python
plt.imshow(img)
````

Assim como com o `img_array`

````python
plt.imshow(img_array)
````

### Converting para RGB

````python
img_rgb = img.convert(mode='RGB')
````

Note que o fundo agora está preto ao convertermos para RGB.

````python
plt.imshow(img_rgb)
````

````python
img_rgb_array = np.array(img_rgb)
````

````python
print(img_rgb_array.shape)
````

### Removendo o fundo preto

Há várias formas de fazermos isso, mas uma das formas é a seguinte

````python
background = Image.new('RGBA', img.size, (255,255,255))
img_rgb2 = Image.alpha_composite(background, img)

img_rgb2_array = np.asarray(img_rgb2)
````

````python
plt.imshow(img_rgb2_array)
````

### Redimensionando a imagem

````python
img_rgb2_resized = img_rgb2.resize((300, 300))
img_rgb2_resized_array = np.array(img_rgb2_resized)
````

````python
plt.imshow(img_rgb2_resized_array)
````

````python
print(img_rgb2_resized_array.shape)
````

### Convertendo para tons de cinza

````python
img_gray = img.convert(mode='L')
````

````python
plt.imshow(img_gray, cmap='gray')
````
